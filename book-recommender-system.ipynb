## Content Filtering Based Recommender System

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
# books.columns
```

```python

books['content'] = books['title'] + ' ' + books['author'] + ' ' + books['genres'] 
```

```python
books = books[books['numRatings'] >= 30000]
books.shape
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books['content'])
```

```python
similarity_scores = cosine_similarity(tfidf_matrix)
```

```python
pickle.dump(popular_df,open('popular.pkl','wb'))
pickle.dump(books, open('books.pkl', 'wb'))
pickle.dump(similarity_scores, open('similarity_scores.pkl', 'wb'))
```

```python
books = pickle.load(open('books.pkl', 'rb'))
similarity_scores = pickle.load(open('similarity_scores.pkl', 'rb'))

def recommend(book_name):
    if book_name not in books['title'].values:
        return f"❌ Book '{book_name}' not found."

    index = books[books['title'] == book_name].index[0]

    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        book_row = books.iloc[i[0]]
        item.append(book_row['title'])
        item.append(book_row['author'])
        item.append(book_row['coverImg'])  # use 'coverImg' instead of 'URL'
        data.append(item)

    return data
```

```python
recommend('Animal Farm')
```
